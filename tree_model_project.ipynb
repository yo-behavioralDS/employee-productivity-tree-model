### Dataset Cleaning II

we need to convert some columns from categorical to numerical to use them in our model.

These are **categorical**: [quarter, day, department, team, productivity]

These are **numerical**:[no_of_workers,targeted_productivity, actual_productivity, over_time, smv, incentive]

Some categorical are stored as numbers, such as quarter and teams, however they represent groups rather than continous values, thus they are categorical. Meanwhile, the other columns contain continuous numeric data

**"department" column**

```python
#code the department: sewing=1, "finishing": 0
df['department']=df['department'].replace({'finishing':0,'sweing':1}).astype('int')
```

```python
df['department'].value_counts()
```

```python
#let's rename the column appropriately
df.rename(columns={'department':'department_code'}, inplace=True)
df.head()
```

**Department codes:**
1. finishing:0
2. sweing:1

**Note: We did not create seperate dummies for this because when there are only two categories, the model doesn’t see them as "ordered steps" — it just sees two groups.**

**"quarter" column**

Because this column is categorical, but is stored in numbers, it could mislead a Decision Tree model into thinking they have numerical order, thus treating it as ordinal or by ranks (quarter 1 is lower than quarter 2).
To prevent this, I will encode each category into seperate dummies, 0s and 1s, having its own columns.

```python
# create the dummies 
dummies=pd.get_dummies(df['quarter'], prefix='q').astype('int') #q_1   q_2   q_3   q_4

#merge the dummies back to df
df=pd.concat([df,dummies],axis=1)

#drop the original column
df.drop(columns=['quarter'],inplace=True)
```

```python
df.sample(10)
```

#### "day" column

Let's repeat for "day" column. Note there are no 'Fridays'

```python
dummies2=pd.get_dummies(df['day'], prefix=None).astype(int)

df=pd.concat([df,dummies2], axis=1)

#drop column
df.drop(columns=['day'], inplace=True)

df.sample(10, random_state = 14)
```

#### "team" column

Same here. There are 12 teams.

```python
#create dummies (12 columns)
dummies3=pd.get_dummies(df['team'], prefix='team').astype(int)

df=pd.concat([df,dummies3], axis=1)

df.drop(columns=['team'],inplace=True)

df.sample(10, random_state = 14)
```

        Completed the Data Cleaning procedure!! Now let's build the tree.

### Tree Model

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

#### Step 1: Define Features and Target
The "productive" column is the **target** (y). Other columns are the **features** (X). 
Do not include the "actual_productivity" because it would be considered cheating.

```python
y=df['productivity']
X=df.drop(columns=['productivity','actual_productivity'], axis=1)
```

```python
X[:3]
```

```python
y[:3]
```

#### STep 2: Split Data into Training and Test Sets
Split data into training and test data:
1. Training: build the model (80% of data)
2. Test: evaluate the accuracy of prediction of new unseen data. (20% of data)

```python
y_train, y_test, X_train, X_test=train_test_split(y, X, test_size=0.2, shuffle= True, random_state=24)
```

#### Step 3: Create and Train the decision tree.
Let's create the tree model and limit its depth to avoid overfitting.

And train the model.

```python
tree=DecisionTreeClassifier(max_depth=3, random_state=24)
tree.fit(X_train, y_train)
```

#### Step 4: Make Predictions and Check Accuracy
Here we use the trained model to make predictions on the X_test set and generate predicted labels for the unseen data.

Then, we compared predictions (y_pred) with the actual labels (y_test) to evaluate model's accuracy.

```python
y_pred=tree.predict(X_test)

accuracy=accuracy_score(y_test, y_pred)
print('Accuracy:', round(accuracy,2))
```

The productive labels was predicted using multiple features with a Decision Tree Classifier. This model achieved 85% accuracy.

### Evaluation I

Evaluate your Decision Tree Classifier using multiple performance metrics — not just accuracy by 1) generating a confusion matrix, 2) Calculate Precision, Recall, and F1 Score, 3)Compare these metrics to Accuracy

```python
#step1 - create confusion matrix

from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, y_pred))
```

Quick visual summary:

|                           |    Predicted **Unproductive** |      Predicted **Productive** |
| ------------------------- | ----------------------------: | ----------------------------: |
| **Actually Unproductive** |  **35** = True Negatives (TN) | **24** = False Positives (FP) |
| **Actually Productive**   | **13** = False Negatives (FN) | **168** = True Positives (TP) |


```python
#step2 - Calculate Precision, Recall, and F1
from sklearn.metrics import precision_score, recall_score, f1_score
print("Precision:", round(precision_score(y_test, y_pred), 2))
print("Recall:", round(recall_score(y_test,y_pred),2))
print("F1 Score:", round(f1_score(y_test,y_pred),2))
```

**Interpretation:**

+ Precision: When the model predicts a line that is productive, it is correct 88% of the time. 
+ Recall: When the model correctly finds 93% of all actual Productive lines, missing very few.
+ F1: The model achieves a 90% balance between precision and recall, meaning it is both accurate and consistent at identifying Productive lines.

Finally, while the model’s overall accuracy is 85%, its precision of 88%, recall of 93%, and F1 score of 90% show that it is even more effective at correctly identifying productive lines than accuracy alone suggests.

```python
# Notes:

# Accuracy = (all correct predictions) ÷ (all predictions)

# Precision = of all predicted positives, how many were correct
# → Focuses on avoiding false positives

# Recall = of all actual positives, how many were caught
# → Focuses on avoiding false negatives

# F1 Score = harmonic mean of precision and recall → balances both
```

### Evaluation II

Now let's see how consistent the model's performance is and not just a lucky result from one split. It will be tested on different multiple splits

```python
from sklearn.model_selection import cross_val_score
```

```python
scores = cross_val_score(tree, X, y, cv=10)

print("Cross Validation Accuracy Scores:", scores.round(2))
print("Cross Validation Accuracy Scores:", scores.mean().round(2))
```

**Interpretation:**

On average, the Decision Tree gets 82% accuracy across all folds, close to the original 85% — this means it’s consistent.

However, accuracy alone is not enough, let's also calculate precision, recall, and F1.

```python
from sklearn.model_selection import cross_validate

multiple_cross_scores= cross_validate(tree, X, y, cv=10, scoring=("precision", "recall", "f1"))
```

```python
print("1. Mean Cross Validated Precision:", round(multiple_cross_scores["test_precision"].mean(),2))
print("2. Mean Cross Validated Recall", round(multiple_cross_scores["test_recall"].mean(),2))
print("3. Mean Cross Validated F1:", round(multiple_cross_scores["test_f1"].mean(), 2))
```

**Interpretation:**
1. On average, when the model predicts "Productive", it’s correct 85% of the time across all folds.
2. The model catches 92% of all actual productive cases consistently across folds.
3. The model has a balanced precision/recall performance of 88%, stable across folds.

**Main Insights:**

As illustrated, the top node(root) represents  the most significant variable in the model, which is 'incentive'. Following along the branches, you'll also see that the variable 'smv' is another significant variable. The model has decided that these two variables are the strongest predictors of whether a team will be productive.

“Incentive” = financial motivation offered

“SMV” = time allocated for a task

**Guideline on reading the tree:**
1) start at the node (the highest point).
2) Each box has a condition like *incentive <= 22.0*
3) Insert the team's actual values and answer *True or False*:
         True → follow the left branch
         False → follow the right branch
4) Repeat this until you land on a leaf node, which gives the prediction (class).

**Let's walk through a hypothetical example:**

Team A:  
1) incentive = 22
2) smv = 4.44

We start at the root of the tree, which checks whether the incentive is less than or equal to 22. Since the input has an incentive of 22, the statement of *22 ≤ 22* being true, so we follow the left branch. Next, we reach a node that checks if the smv is less than or equal to 3.92. We have an smv of 4.44, which deems the statement of *4.44 <= 3.92* false, so we follow the rght branch. Finally, *4.44 ≤ 4.865* is True, so we follow the left branch and arrive at a leaf node which gives the prediction of the team being "Productive".

**Final Insight:**
We can now predict productivity and understand exactly what drives it — giving us actionable insight to improve performance.

### Random Forest

As a final validation step, a Random Forest will be conducted due to a Decision Tree’s tendency to overfit the data.

```python
from sklearn.ensemble import RandomForestClassifier #builds multiple decision trees on random subsets of the data and averages their predictions to get a stronger one.
```

```python
forest = RandomForestClassifier(
    oob_score = True,
    random_state = 24
)

forest.fit(X_train, y_train) #trains the forest on training data

y_pred_forest = forest.predict(X_test) #make predictions on the test set

acc=accuracy_score(y_test, y_pred_forest) #compares the forests predictions with the real test labels.

print("Accuracy:", round(acc,2))
```

Now we check the Out of Bag Score, which is the forest’s internal accuracy estimate from data each tree didn’t see during training.

```python
print("Out Of Bag Score:", round(forest.oob_score_, 2))
```

Finally, the evaluation metrics of the Random Forest are similar to those of the Decision Tree, which further reconfirms the reliability of our model’s results.
